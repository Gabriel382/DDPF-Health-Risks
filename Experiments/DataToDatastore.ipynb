```python
from gatenlp import Document
from gatenlp.gateworker import GateWorker
from gatenlp.corpora import ListCorpus
import glob
from tqdm import tqdm
```

```python
path_to_covid_social = "../data/covid/social/documents/2020-2/"
```

```python
gs = GateWorker(start=False, auth_token="1234")
```

```python
def createCorpusFromFolder(gs,corpusName, pathToFolder):
    corpus = gs.worker.newCorpus()
    corpus.setName(corpusName)
    for docpath in tqdm(glob.glob(pathToFolder+'*.txt')):
        doc = Document.load(docpath)
        pdoc = gs.pdoc2gdoc(doc)
        corpus.add(pdoc)
        gs.del_resource(pdoc)
    return corpus
```

```python
# COVID Reddit
path_to_covid_social = "../data/covid/social/documents/2020-2/"
corpus = createCorpusFromFolder(gs,'PreDiViD-COVID-2022-02',path_to_covid_social)
```
